```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
```

```python
# Exercise 1: estimate
e1_n = [1, 2, 5, 10]

for i in e1_n:
  # estimate
  e = (1 + (1/i))**i
  
  # np.exp(1) is also an estimate but it is going to be much more accurate than
  # our calculated e hence we will use it
  print(f'e1_n: {i:2.0f}, est.e: {e:.5f}, diff to e: {(np.exp(1) - e):.7f}')
```

```python
# Exercise 2: Visualize e's approach
e2_n = np.arange(1, 1001)

# numpy arr is different from simple list(used above)
# on numpy arr (vector n)
# element wise algebric operations (divison, power) can be perfomed
# 
e_diffs = np.exp(1) - (1 + 1/e2_n)**e2_n
plt.plot(e2_n, e_diffs)
```

```python
# Exercise 3: Exploring e in numpy
x_domain = [-2, 2]
x = np.linspace(x_domain[0], x_domain[1], 53)
x
```

```python
# functions
y1 = np.exp(x)
y2 = np.exp(x**2)
y3 = np.exp((-x)**2)
y4 = np.exp(-(x**2))
y5 = np.exp(x)**2

# and plot
plt.figure(figsize=(8,6))
plt.plot(x, y1, linewidth=2, label='$y=e^x$')
plt.plot(x, y2, linewidth=2, label='$y=e^{x^2}$')
plt.plot(x, y3, '--', linewidth=2, label='$y=e^{(-x)^2}$')
plt.plot(x, y4, linewidth=2, label='$y=e^{-(x^2)}$')
plt.plot(x, y5, linewidth=2, label='$y={(e^x)}^2$')

plt.xlim([x[0], x[-1]])
plt.ylim([-1, 10])

plt.xlabel('x')
plt.ylabel('y=f(x)')

plt.legend()
plt.grid()
plt.show()
```

```python
# Exercise 4: Exploring e in sympy
s_beta = sym.var('beta')

s_y = sym.exp(s_beta) - sym.log(s_beta) - np.exp(1)
s_y

sym.plot(s_y, (s_beta, x_domain[0], x_domain[1]), title=f'$f(\\beta) = {sym.latex(s_y)}$', ylabel='$y=f(\\beta)$', xlabel='$\\beta$');
```

```python
# Exercise 5: exp and log
e5_ds = np.linspace(-4, 4, 30)
plt.plot(e5_ds, np.exp(e5_ds), linewidth=2, label='exp(x)')

plt.plot(e5_ds, np.log(np.exp(e5_ds)), linewidth=2, label='log(exp(x))')
plt.plot(e5_ds, np.exp(np.log(e5_ds)), 'o', linewidth=2, label='exp(log(x))')


# log with fine grained dataset
e5_ds_for_log = np.linspace(.001, 4, 30)
plt.plot(e5_ds_for_log, np.log(e5_ds_for_log), linewidth=2, label='log(x)')



plt.legend()
# plt.axis('square')
plt.xlim([-4, 4])
plt.ylim([-4, 4])
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
