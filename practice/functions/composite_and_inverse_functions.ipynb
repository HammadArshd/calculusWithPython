```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
```

```python
#  Exercise 1a: Composite functions in numpy
```

```python
def fx(x):
  return 2 * x**2 - 4

def gx(x):
  return 7*np.abs(x) + 3
```

```python
def fgx(xx):
  return fx(gx(xx))

def gfx(xx):
  return gx(fx(xx))
```

```python
xx = np.linspace(-5, 5, 200)
xx
```

```python
plt.plot(xx, fx(xx), label='f(x)')
plt.plot(xx, gx(xx), label='g(x)')
plt.plot(xx, fgx(xx), label='f(g(x))')
plt.plot(xx, gfx(xx), '.', label='g(f(x))')
plt.ylim([-5, 50])
plt.legend()
plt.show()
```

```python
# Exercise 1b: Composite functions in numpy
```

```python
def fx(x):
  return np.sin(x)

def gx(x):
  return np.log(x)

def hx(x):
  return 2*x**2 + 5
```

```python
def fghx(x):
  return fx(gx(hx(x)))
```

```python
xx = np.linspace(-100*np.pi, 100*np.pi, 1001)
plt.plot(xx, fghx(xx), label='f(g(h(x)))')
plt.legend()
plt.show()
```

```python
# Exercise 2: Inverse function in numpy
```

```python
def f(x):
  return np.log(2*x)

def g(x):
  return np.exp(x) / 2
```

```python
xx = np.linspace(.01, 5, 54)
plt.plot(xx, f(xx), label='f(x)')
plt.plot(xx, g(xx), label='g(x)')
plt.plot(xx, f(g(xx)), label='f(g(x))')
plt.plot(xx, g(f(xx)), '.', label='g(f(x))')
# plt.xlim([1, 5])
plt.ylim([-5, 10])
plt.legend()
plt.show()
```

```python
# Exercise 3: Composite functions in sympy
```

```python
sx = sym.symbols('x')
fx = sym.sin(sx)
gx = sym.log(sx)
hx = 2*sx**2 + 5
fx, gx, hx
```

```python
# use of .subs
fx.subs(sx, 0)
```

```python
# original fx
fx
```

```python
fghx = fx.subs(sx, gx.subs(sx, hx))
fghx
```

```python
sym.plot(
    fghx,
    (sx, -100, 100),
    title=f'$y = {sym.latex(fghx)}$'
)
```

```python
# Exercise 4: Inverting functions in sympy
```

```python
sy = sym.symbols('y')

fx = 2*sx + 3
gx = 2*sx + sym.sin(sx)

display(fx)
display(gx)
```

```python
# Solve for inverse
invfx = sym.solve(sy - fx, sx)[0]
# invgx = sym.solve(sy - gx, sx)[0] # non invertible function
display(invfx)
# display(invgx)
```

```python
fx.subs(sx, 4)
```

```python
invfx.subs(sy, (fx.subs(sx, 4)))
```

```python
fx.subs(sx, invfx.subs(sy, 4))
```
