```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
```

```python
# Exercise 1: Fun with trig
xx = np.linspace(-np.pi, 2*np.pi, 1234)
xx
```

```python
f = [0]*3
f[0] = np.sin(xx), 'sin(x)'
f[1] = np.sin(xx)**2, 'sin(x)**2'
f[2] = np.sin(xx**2), 'sin(x**2)'

for y,label in f:
  plt.plot(xx, y, label=label)

plt.legend()
plt.xlabel('Angle (rad.)')
plt.ylabel('y = f(x)')
plt.show()
```

```python
# Exercise 2: More fun with trig
xx = np.linspace(-np.pi, 2*np.pi, 1234)
xx
```

```python
f = [0]*3
f[0] = np.sin(np.cos(xx)),  'sin(cos())'
f[1] = np.cos(np.sin(xx)), 'cos(sin())'
f[2] = np.cos(xx), 'cos()'

for y,label in f:
  plt.plot(xx, y, label=label)

plt.legend()
plt.xlabel('Angle (rad.)')
plt.ylabel('y = f(x)')
plt.show()
```

```python
# Exercise 3: resolution and tan
for p in range(2, 5):
  dataset = np.linspace(0, 2*np.pi, 10**p)
  plt.plot(dataset, np.tan(dataset), label=f'{10**p} pnts')

plt.legend()
plt.xlabel('Angle (rad.)')
plt.ylabel('y = f(x)')
plt.show()
```

```python
# Exercise 3: resolution and tan --- zoomed in
for p in range(2, 5):
  dataset = np.linspace(0, 2*np.pi, 10**p)
  plt.plot(dataset, np.tan(dataset), label=f'{10**p} pnts')

plt.xlim([1.55, 1.6])
plt.legend()
plt.xlabel('Angle (rad.)')
plt.ylabel('y = f(x)')
plt.show()
```
