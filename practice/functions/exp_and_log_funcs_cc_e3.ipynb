```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
```

```python
x_domain = [-2, 2]
x = np.linspace(x_domain[0], x_domain[1], 53)
x

# functions
y1 = np.exp(x)
y2 = np.exp(x**2)
y3 = np.exp((-x)**2)
y4 = np.exp(-(x**2))
y5 = np.exp(x)**2

# and plot
plt.figure(figsize=(8,6))
plt.plot(x, y1, linewidth=2, label='$y=e^x$')
plt.plot(x, y2, linewidth=2, label='$y=e^{x^2}$')
plt.plot(x, y3, '--', linewidth=2, label='$y=e^{(-x)^2}$')
plt.plot(x, y4, linewidth=2, label='$y=e^{-(x^2)}$')
plt.plot(x, y5, linewidth=2, label='$y={(e^x)}^2$')

plt.xlim([x[0], x[-1]])
plt.ylim([-1, 10])

plt.xlabel('x')
plt.ylabel('y=f(x)')

plt.legend()
plt.grid()
plt.show()
```
