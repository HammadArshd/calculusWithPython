```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
```

```python
np.set_printoptions(suppress=True)
```

```python
xDomain = [-6, 6]
resolution = .5
n_coefs = np.random.randn(4)
x = np.arange(xDomain[0], xDomain[1]+resolution, resolution)
y = np.zeros(len(x))
fname = '$y = '
for i,c in enumerate(n_coefs):
  y += c*x**i
  fname += '+ '[int(c<0)] + f'{c:.2f}x^{i}'
y, n_coefs
```

```python
plt.plot(x, y)
plt.title(fname + '$')
plt.grid()
plt.xlim(xDomain[0], xDomain[1])
plt.xlabel('x')
plt.ylabel('y=f(x)')
plt.show()
```
