```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
from decimal import getcontext, Decimal
```

```python
xDomain = [-2, 2]

#Dealing with floats and Decimals
df_test = np.arange(Decimal('-2.0'), Decimal('2.0'), Decimal('0.1'))
df_test
```

```python
# Option 1
resolution = .1
x = np.arange(xDomain[0], xDomain[1]+resolution, resolution)
np.set_printoptions(suppress=False)
x
```

```python
np.set_printoptions(suppress=True)
x
```

```python
# Option 2
x2 = np.linspace(xDomain[0], xDomain[1], 41)
# ODD number above : 41 will gurentee a true 0
x2
```

```python
# Function - in form of math function and not as python function
y = x**2 + 3*x**3 - x**4
y
```

```python
# plot
plt.plot(x, y, label='$y=x^2+3x^3-x^4$')
plt.grid()
plt.legend()
plt.xlim(xDomain[0], xDomain[1])
plt.xlabel('x')
plt.ylabel('y=f(x)')
plt.show()
```

```python
# symbolic variable
s_beta = sym.var('beta')
s_beta
# xN or n_ or -> numpy variable
# xS or s_ or -> sympy variable

# function
s_y = s_beta**2 + 3*s_beta**3 - s_beta**4
s_y
```

```python
sym.plot(s_y,(s_beta, xDomain[0], xDomain[1]), xlabel='x', ylabel=None, title=f'$f(\\beta)={sym.latex(s_y)}$')
```

```python
# converting sympy objects into functions that we can query and use in numpy and matplotlib 
fx = sym.lambdify(s_beta, s_y)
fx(2)

l_y = fx(x)
plt.plot(x, l_y)
```
