```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
```

```python
# Exercise 4: Evaluate the function at or near a point
```

```python
# with numpy
xx = np.linspace(-3, 5, 100)    # rought approximation

# increase resolution for better approximation
#xx = np.linspace(-3, 5, 1000)   # better approximation
#xx = np.linspace(-3, 5, 10000)  # even better approximation
piece1 = 0 * (xx < 0)
piece2 = -2 * xx * ((xx >= 0) & (xx<3))
piece3 = xx**3/10 * (xx >= 3)
y = piece1 + piece2 + piece3
xx
```

```python
xloc = .5
shifted_vector_xx = xx - xloc
# shift the value, and in new values required value is closet to 0
shifted_vector_xx
```

```python
plt.plot(shifted_vector_xx)
```

```python
# approach, take the abs value and look for minimum
plt.plot(np.abs(shifted_vector_xx))
```

```python
# another approach, square the values, which has same effect
plt.plot(shifted_vector_xx**2)
```

```python
# minimum of the function shifted_vector_xx**2 corresponds to the point in,
# x-axis that is close to the .5
xidx = np.argmin(shifted_vector_xx**2)
xidx
```

```python
# possible closet number, in vector xx, to the desired number .5
xx[xidx]
```

```python
y[xidx]
```

```python
# with sympy
from sympy.abc import x
piece1 = 0
piece2 = -2*x
piece3 = x**3/10
fx = sym.Piecewise(
    (piece1, (x < 0)),
    (piece2, (x >= 0) & (x < 3)),
    (piece3, (x >= 3))
)
fx.subs(x, xloc)
# is giving direct and complete answer
```

```python
# Difference: why its diffcult to locate specific point in numpy than sympy

# sympy is encoding the piecewise fucntion as symbolic expression
# whereas
# numpy is storing computed numbers 
```
